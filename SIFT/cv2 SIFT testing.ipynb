```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('SIFT_testing_image.PNG')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
sift = cv.SIFT_create()
keypoints = sift.detect(gray, None)
print("Number of keypoints: ", len(keypoints))
```

```python
img = cv.drawKeypoints(gray, keypoints, img)

fig, axes = plt.subplots(1, 2, figsize=(12, 10))
ax = axes.ravel()
ax[0].imshow(gray)
ax[0].set_title("Gray Scale Image")
ax[1].imshow(img)
ax[1].set_title("Keypoints on Image")
plt.show()
```

```python
keypoints, descriptor = sift.compute(gray, keypoints)
```

```python
# Now we have our descriptor for each keypoint:
print(descriptor[1480])
```
